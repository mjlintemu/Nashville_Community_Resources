# Cleaning Public Dataset

```python
# Filtering data down to stateCode TN
public_tn = public[public['stateCode']=='TN']
public_tn.head()
```

```python
# Filter County to Davidson
public_davidson = public_tn[public_tn['county']=='Davidson']
public_davidson.head()
```

```python
# Drop Columns
public_davidson_2 = public_davidson.drop(['stateNumberCode','hash','lastRefresh','id'], axis=1)
public_davidson_2.head()
```

```python
# adding column for declartion date as a datetime datatype
public_davidson_2['declaration_date'] = pd.to_datetime(public_davidson_2['declarationDate'])
public_davidson_2.head()
```

```python
# adding column for obligated date as a datetime datatype
public_davidson_2['obligated_date'] = pd.to_datetime(public_davidson_2['obligatedDate'])
public_davidson_2.head()
```

```python
# Drop unncessary columns
public_davidson_3 = public_davidson_2.drop(['declarationDate','pwNumber','damageCategoryCode','obligatedDate'], axis=1)
public_davidson_3.head()
```

```python
public_clean = public_davidson_3.rename(columns={"disasterNumber": "Disaster_Number", "incidentType": "Incident_Type", 'applicationTitle':'Application_Title','applicantId':'Applicant_Id','dcc':'DCC','damageCategory':'Damage_Category','projectSize':'Project_Size','county':'County','countyCode':'County_Code','state':'State','stateCode':'State_Code','projectAmount':'Project_Amount','federalShareObligated':'Federal_Share_Obligated','totalObligated':'Total_Obligated'})
public_clean.head()
```

```python
# Drop unncessary columns
public_clean_2 = public_clean.drop(['Applicant_Id','County_Code','State'], axis=1)
public_clean_2
```

```python
public_clean_3 = public_clean_2[public_clean_2["Incident_Type"].str.contains("Biological")==False]
public_clean_3
```

```python
public_project = public_clean_3
public_project
```

# Cleaning Owners Dataset

```python
owners.head()
```

```python
# Filtering data down to State Code TN
owners_tn = owners[owners['state']=='TN']
owners_tn.head()
```

```python
# Cleaning 'county' column to remove (County)
owners_tn['county'] = owners_tn['county'].str.replace(r'\(.*\)','',regex=True)
owners_tn
```

```python
# Filter dataset to Davidson county. Having issues filtering by county so I used city.
owners_davidson = owners_tn[owners_tn['city']=='NASHVILLE']
owners_davidson.head()
```

```python
# Drop Columns
owners_davidson_2 = owners_davidson.drop(['validRegistrations','approvedBetween1And10000','approvedBetween25001AndMax','totalMaxGrants','id'], axis=1)
owners_davidson_2.head()
```

```python
# Drop Columns
owners_davidson_3 = owners_davidson_2.drop(['noFemaInspectedDamage','femaInspectedDamageBetween1And10000','femaInspectedDamageBetween10001And20000','femaInspectedDamageBetween20001And30000','femaInspectedDamageGreaterThan30000','approvedForFemaAssistance','approvedBetween10001And25000'], axis=1)
owners_davidson_3.head()
```

```python
owners_davidson_4 = owners_davidson_3.rename(columns={"disasterNumber": "Disaster_Number", "state": "State", 'county':'County','city':'City','zipCode':'Zip_Code','averageFemaInspectedDamage':'Avg_Inspected_Damage','totalInspected':'Total_Inspected','totalDamage':'Total_Damage','totalApprovedIhpAmount':'Total_Approved','repairReplaceAmount':'Repair_Replace_Amount','rentalAmount':'Rental_Amount','otherNeedsAmount':'Other_Needs_Amount'})
owners_davidson_4
```

```python
owners_davidson_4.County.unique()
```

```python
owners_davidson_5 = owners_davidson_4[owners_davidson_4['County']=='Davidson ']
owners_davidson_5
```

```python
owners_project = owners_davidson_5.drop(['City'], axis=1)
owners_project
```
